```python
# Bar Diagram

import pandas as pd
import matplotlib.pyplot as plt
url='https://raw.githubusercontent.com/ridoy365/Ridoy-01-PP-07/refs/heads/main/environmental_data.csv'
df=pd.read_csv(url)

temperature_sum=df['Temperature'].sum()
humidity_sum=df['Humidity'].sum()
pressure_sum=df['Pressure'].sum()

y=[temperature_sum,humidity_sum,pressure_sum]
x=["Temperature","Humidity","Pressure"]

plt.bar(x,y, color='orange')

plt.grid(color='black',linestyle='--',linewidth='.55')

plt.title("Distribution of Environmental Data")

plt.xlabel("Variables")

plt.ylabel("Sum of Values")

plt.show()
```

```python
# Pie Chart

import pandas as pd
import matplotlib.pyplot as plt
url='https://raw.githubusercontent.com/ridoy365/Ridoy-01-PP-07/refs/heads/main/environmental_data.csv'
df=pd.read_csv(url)

temperature_sum=df['Temperature'].sum()
humidity_sum=df['Humidity'].sum()
pressure_sum=df['Pressure'].sum()

data=[temperature_sum,humidity_sum,pressure_sum]
labels=["Temperature","Humidity","Pressure"]

plt.pie(data,labels=labels,autopct='%1.1f%%', startangle=90, radius=1, explode=[.1,0,0])

plt.grid(color='black',linestyle='--', linewidth='.50')

plt.title("Distribution of Environmental Data")

plt.show()
```

```python
# Line Graph

import pandas as pd
import matplotlib.pyplot as plt
url='https://raw.githubusercontent.com/ridoy365/Ridoy-01-PP-07/refs/heads/main/environmental_data.csv'
df=pd.read_csv(url)

plt.plot(
    df["Humidity"],
    df["Temperature"],
    label="Temperature and Humidity", color="green", ls="dotted", marker="o"
    )

plt.plot(
    df["Pressure"],
    df["Temperature"],
    label="Temperature and Pressure", color="red", ls="solid", marker="o"
    )

plt.grid(color='black', linestyle='-.', linewidth='0.60')

plt.xlabel("Humidity                                                  Pressure")

plt.ylabel("Temperature")

plt.title("Temperature-Humidity and Temperature-Pressure")

plt.show()
```

```python
# Use of Seaborn

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
url='https://raw.githubusercontent.com/ridoy365/Ridoy-01-PP-07/refs/heads/main/environmental_data.csv'
df=pd.read_csv(url)

temperature_sum=df['Temperature'].sum()
humidity_sum=df['Humidity'].sum()
pressure_sum=df['Pressure'].sum()

data=[temperature_sum, humidity_sum, pressure_sum]

labels=["Temperature","Humidity","Pressure"]

df=pd.DataFrame({'factor':x,'sum':data})

sns.set(style="whitegrid")

sns.set(palette="muted")

sns.barplot(x="factor", y="sum", data=df, color="brown")

plt.title("Distribution of Environmental Factors", fontsize=12, fontweight="bold", color="black")

plt.xlabel("Variables", fontsize=10)

plt.ylabel("Sum of Values", fontsize=10)

plt.show()
```

```python
#Bokeh

import pandas as pd
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import HoverTool
url='https://raw.githubusercontent.com/ridoy365/Ridoy-01-PP-07/refs/heads/main/environmental_data.csv'
df=pd.read_csv(url)

temperature=df["Temperature"]
humidity=df["Humidity"]
pressure=df["Pressure"]

output_notebook

p=figure(title="Temperature-Humidty and Temperature-Pressure", x_axis_label="Humidity/Pressure", y_axis_label="Temperature")

p.line(humidity, temperature, line_width=3, color="yellow", legend_label="Temperature-Humidity")

p.scatter(humidity, temperature, size=11, color="red", legend_label="Temperature-Humidity")

hover=HoverTool()

hover.tooltips=[("x","$x"),("y","$y")]

p.add_tools(hover)

show(p)
```
